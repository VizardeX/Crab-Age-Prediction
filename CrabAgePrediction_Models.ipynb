```python
train_df = train_df.drop(['Shell Weight'], axis=1)
train_df.head()
```

```python
#we need to change the values in the Sex column to numerical values to use it in the ML model
train_df['Sex'] = train_df['Sex'].map({ 'I':0, 'M':1, 'F':2 })
train_df.head()
```

```python

y = train_df['Age']
x = train_df.drop(['id', 'Age'], axis=1)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
x.head()
```

```python
from sklearn.linear_model import Lasso

lasso_model = Lasso(alpha=1, max_iter=100, tol=0.1)
lasso_model.fit(x_train, y_train)
lasso_pred =lasso_model.predict(x_test)

mean_absolute_error(y_test, lasso_pred)
```

```python
from sklearn.linear_model import Ridge
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import PolynomialFeatures

ridge_model = Pipeline([
    ('PolynomialFeatures', PolynomialFeatures(degree=4)),         #used PolynomialFeatures to enhances the interctions between the features
    ('Ridge Regression', Ridge(alpha=2, max_iter=100, tol=0.1))
])

ridge_model.fit(x_train, y_train)

ridge_pred = ridge_model.predict(x_test)
mean_absolute_error(y_test, ridge_pred)
#The Ridge regression model has the best mae, so we will enhance it further
```

```python
from sklearn.linear_model import ElasticNet

elastic_model = ElasticNet(alpha=1, l1_ratio=0.5, max_iter=100, tol=0.1)
elastic_model.fit(x_train, y_train)

elastic_pred = elastic_model.predict(x_test)

mean_absolute_error(y_test, elastic_pred)
```

```python
from xgboost import XGBRegressor

xg_model = XGBRegressor(n_estimators=25, learning_rate=0.1, max_depth=3)
xg_model.fit(x_train, y_train)

xg_pred = xg_model.predict(x_test)

mean_absolute_error(y_test, xg_pred)
```
